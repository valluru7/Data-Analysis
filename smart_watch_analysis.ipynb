```python
#convert date datatype from object to datetime
#data["ActivityDate"] =  pd.to_datetime(data["ActivityDate"], format="%m/%d/%y")
data["ActivityDate"] =  pd.to_datetime(data["ActivityDate"], format="%m/%d/%y", exact = False)
#data['ActivityDate'] = datetime.strptime(data['ActivityDate'],"%d %B, %Y")
data.info()
```

```python
data['ActivityDate'].isna().sum()
```

```python
data['ActivityDate'].isna().values.any()
```

```python
data['TotalMinutes'] = data['VeryActiveMinutes']+data['FairlyActiveMinutes']+data['LightlyActiveMinutes']+data['SedentaryMinutes']
data['TotalMinutes'].sample(5)
```

```python
#analysing the relation between calories and total steps walked
plot = px.scatter(data_frame = data,
                 y = 'TotalSteps',
                 x = 'Calories',
                 trendline = 'ols')
plot.show()
```

```python
label = ["Very Active Minutes", "Fairly Active Minutes","Lightly Active Minutes", "Inactive Minutes"]
counts = data[["VeryActiveMinutes", "FairlyActiveMinutes", 
         "LightlyActiveMinutes", "SedentaryMinutes"]].mean()
colors = ['gold','lightgreen', "pink", "red"]
fig = go.Figure(data=[go.Pie(labels = label, values = counts)])
fig.update_layout(title_text = 'Total Minutes')
fig.update_traces(hoverinfo='label+percent', textinfo='value', textfont_size=25,
                  marker=dict(colors=colors, line=dict(color='black', width=3)))
fig.show()
```

```python
data['day']=data['ActivityDate'].dt.day_name()
data['day'].head()
```

```python
#let’s have a look at the very active, fairly active, and lightly active minutes on each day of the week:
fig = go.Figure()
fig.add_trace(go.Bar(
        y= data['VeryActiveMinutes'],
        x = data['day'],
        name = 'Very Active',
        marker_color = 'purple'))
fig.add_trace(go.Bar(
        x = data['day'],
        y = data['FairlyActiveMinutes'],
        name = 'Fairly Active',
        marker_color = 'green'))
fig.add_trace(go.Bar(
        x = data['day'],
        y = data['LightlyActiveMinutes'],
        name = 'Lightly Active',
        marker_color = 'blue'))
fig.update_layout(barmode='group', xaxis_tickangle = -45)
fig.show()
```

```python
#inactive minutes in each day of the week
day = data['day'].value_counts()
label = day.index
counts = data['SedentaryMinutes']
colors = ['gold','lightgreen', "pink", "blue", "skyblue", "cyan", "orange"]
fig = go.Figure(data=[go.Pie(labels=label, values=counts)])
fig.update_layout(title_text='Inactive Minutes Daily')
fig.update_traces(hoverinfo='label+percent', textinfo='value', textfont_size=30,
                  marker=dict(colors=colors, line=dict(color='black', width=3)))
fig.show()
```

```python
#number of calories burned
calories = data['day'].value_counts()
label = calories.index
counts = data['Calories']
color = ['gold','lightgreen', "pink", "blue", "skyblue", "cyan", "orange"]
fig = go.Figure(data=[go.Pie(labels = label, values = counts)])
fig.update_layout(title_text = 'Calories Burned')
fig.update_traces(hoverinfo = 'label+percent', textinfo = 'value', textfont_size = 30,
                 marker=dict(colors = color, line=dict(color = 'black', width = 3)))
fig.show()
```
